```python
import os
import matplotlib.pyplot as plt
import pandas

root = os.getcwd()
file = os.path.join(root, 'Downloads', 'company_sales_data.csv')
file
```

```python
data = pandas.read_csv(file)
data.head()
```

```python
#Прочитайте все данные о продажах продукта и покажите их с помощью многострочного графика

data_func = data.copy()
del data_func['month_number']
del data_func['total_units']
del data_func['total_profit']

month_number = data['month_number'].tolist()

f = pandas.DataFrame(data_func)
f.plot(kind = 'bar')
plt.title('Месячные данные о продажах')
plt.xlabel('Месяца')
plt.ylabel('Объём продаж')
plt.xticks(month_number)
plt.show()
```

```python
#Прочитайте данные о продажах мыла для купания за все месяцы и покажите их с помощью столбчатой диаграммы.

bathingsoap = data['bathingsoap'].tolist()
month_number = data['month_number'].tolist()

plt.bar(month_number, bathingsoap)
plt.title('Продажи мыла для купания')
plt.xlabel('Месяца')
plt.ylabel('Объём продаж')
plt.xticks(month_number)
plt.show()
```

```python
#Рассчитайте общие данные о продажах за год для каждого продукта и покажите их с помощью круговой диаграммы

total = []
name_produckt = []
for key in data:
    buf = sum(data[key].tolist())
    name_produckt.append(key)
    total.append(buf)

colors = ['yellow', 'green', 'red', 'blue', 'orange', 'purple']
plt.pie(total[1:7], labels=name_produckt[1:7], colors=colors,shadow=True,autopct='%1.1f%%',startangle=180)
plt.axis('equal')
plt.show()
```
